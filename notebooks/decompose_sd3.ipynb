```python
from teamwork.pipelines import TeamworkPipeline
from PIL import Image
from IPython import display
import torch
```

```python
pipe = TeamworkPipeline.from_checkpoint(
    'samsartor/teamwork-release',
    'decomposition_heterogeneous_sd3.safetensors',
    torch_dtype=torch.bfloat16,
).to('cuda')
```

```python
image = Image.open('../demo/red_glass_sphere.png')
image.resize([256, 256])
```

```python
generated = pipe({'image': image})
```

```python
for name, out in generated.items():
    display.display(display.HTML(f'<h2>{name}</h2>'))
    display.display(out.resize([256, 256]))
```
